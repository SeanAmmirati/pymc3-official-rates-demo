# Simplistic model of mis-read rate

Assume that:

$$
p \sim \text{Normal}(\alpha \mu, \sigma^2)\\
t \sim \text{Normal}(\mu, \tau^2)\\
(\alpha - 5) / 8 \sim \text{Gamma}(5)
$$

There's a lot of greek characters up there. 

You may be wondering about the strange definition of alpha. I just did this so the distribution would like the distribution below, which is based on the parametization of this from scipy.stats. Just think of this as Gamma "moved over" to the right and spread out a bit. 

The greek characters need some parametizations too. Let's make it really large to account for the lack of knowledge here (I personally have no sense how variable the true and "predicted" numbers may actually be...) Practically, I'll just choose arbitrarily large numbers for the prior...


$$
\sigma \sim HalfNormal(1,000,000)\\
\tau \sim HalfNormal(10,000)\\
\mu \sim Normal(10,000, 10,000) 
$$

Of course, $\mu$ is non sensible here. It can't be negative! Take this to be an uninformed prior -- this can be refined a bit more. Right now I'm essentially saying "this can be anything...but I expect somewhere around 10000...ish...kind of)

This is a bit more of an art than science. Such why priors are such a hot topic.

Because we had no observations, this is really more similar to a prior than a posterior...

So let's try with some observations instead. 

I'm just going to make these up as follows -- the "true testing" number is 700, the "test sd" is 100. 


```python
n = 1000
actual_alpha = 90

number_tested = 1000

true_test = 700
test_sd = 100
true_actual = true_test * actual_alpha
actual_sd = test_sd * 13**2

people_test_positive = np.random.normal(true_test, test_sd, n) 
people_actual_positive = np.random.normal(true_actual, actual_sd, n)

observed = pd.DataFrame({'p': people_test_positive, 't': people_actual_positive})
```

```python
for i in [10 ** i for i in range(4)]:
    print(f'After {i} Observations:')
    use = observed.iloc[0:i]
    
    with pm.Model() as model:
        tau = pm.HalfNormal('tau', 5000)
        sigma = pm.HalfNormal('sigma', 500000)
        mu = pm.Normal('mu', 0, 1000)
    

        beta = pm.Gamma('beta', alpha=5, beta=1/8)
        alpha = pm.Deterministic('alpha', (beta - 5))
        t = pm.Normal('t', alpha * mu, sigma, observed=use['t'])
        p = pm.Normal('p', mu, tau, observed=use['p'])

        trace = pm.sample()
        pm.plot_posterior(trace)
        plt.show()
```
